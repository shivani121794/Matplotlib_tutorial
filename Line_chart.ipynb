```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Basic Graphs

# p1 = (1,4) ; p2 = (2,5) ; p3 = (3,6)

plt.plot([1,2,3,5,7], [4,5,6,1,3])
plt.show()
```

```python
x = [0,1,2,3]
y = [0,4,5,6]
plt.scatter(x,y,marker='*', color='r')
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
x = [1,2,3]
y = [3,6,9]
plt.scatter(x,y,marker='*', color='g')
plt.plot(x,y)
plt.title("Graph between height and weight of mice")
plt.xlabel("height of the mice")
plt.ylabel("weight of the mice")
plt.grid()
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y,label='y=2x',color='y',linewidth=3,marker='*',markersize=10,markeredgecolor='blue')
plt.plot(y,x,label='y=x/2',color='b',linewidth=2,marker='o',markersize=10,markeredgecolor='r',linestyle='--')
plt.plot(x,x,label='x=x',color='g',linewidth=2,marker='o',markersize=6,markeredgecolor='blue',linestyle='--')

plt.title("Multiple Graphs")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xticks([0,1,2,3,4,5,6,7,8])
plt.yticks([0,1,2,3,4,5,6,7,8])
plt.grid()
plt.legend()
plt.show()
```

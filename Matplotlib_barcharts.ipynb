```python
#importing the basic libraries

import numpy as np
import pandas as pd

```

```python
# lets read the dataset

data = pd.read_csv('employee.csv')

# Lets take only 200 rows
data = data.loc[:199,:]
print(data.shape)
```

```python
#BAR PLOT

# Lets use the values in x to plot

plt.figure(figsize=(10,9))
data['Attrition'].value_counts().plot(kind='bar', color='green')

plt.title('Univariate Analysis on Attrition column of Employee Dataset')
plt.xlabel('Attrition')
plt.ylabel('Values')
plt.show()
```

```python
# Analyzing the attribute Department

data['Department'].value_counts().plot(kind='bar', color='lightblue')

plt.title('No of employee in each Departments', fontsize=12)
plt.xlabel('Different Departments')
plt.ylabel('Values')
#plt.xticks()
plt.show()
```

```python
data['Department'].value_counts().plot(kind='bar', color='lightblue')

plt.title('No of employee in each Departments', fontsize=12)
plt.xlabel('Different Departments')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyzing the education field

plt.figure(figsize=(8,4))
data['EducationField'].value_counts().plot(kind='bar', color='orange')
plt.title('No of employee from different Education Fields', fontsize=12)
plt.xlabel('Different Education Fields')
plt.ylabel('Values')
plt.xticks(rotation=90)
plt.show()
```

```python
data.groupby("EducationField")['MonthlyRate'].mean().plot(kind="bar")
```

```python
#Multivariate Bars

data.groupby(["Department", "EducationField"])["MonthlyRate"].mean().plot(kind="bar")
plt.xticks(rotation=90)
plt.show()
```

```python
#Crosstab - Multi pivot table

x = pd.crosstab(index = data['Department'],
                columns = data['EducationField'],
                values = data['MonthlyRate'],
                aggfunc = 'mean')

x
```

```python
#Stacked bar chart

x.plot(kind='bar', stacked=True)
plt.ylabel('Salary')
plt.xticks(rotation=20)
plt.show()
```

```python
x.plot(kind='bar', stacked=False)
plt.ylabel('Salary')
plt.xticks(rotation=20)
plt.show()
```

```python
x.plot(kind='bar', stacked=True)
plt.ylabel('Salary')
plt.xticks(rotation=20)
plt.legend()
plt.show()
```

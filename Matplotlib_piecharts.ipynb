```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('employee.csv')
```

```python
data = data.loc[:199,:]
print(data.shape)
```

```python
import matplotlib.pyplot as plt

plt.pie(data['BusinessTravel'].value_counts(),
        labels = data['BusinessTravel'].value_counts().index, autopct='%1.2f%%')
plt.show()
```

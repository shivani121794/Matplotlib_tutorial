```python
# Import the dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import Seaborn

import seaborn as sns
```

```python
#in seaborn there is already inbuilt dataset

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
print(sns.get_dataset_names())
```

```python
current_palette1 = sns.color_palette()

sns.palplot(current_palette1)

plt.show()
```

```python
# current_palette2 = sns.color_palette()

sns.palplot( sns.color_palette("Greens"))

plt.show()
```

```python
# current_palette3 = sns.color_palette()

sns.palplot(sns.color_palette("BrBG",8))   #8 cells

plt.show()
```

```python
# Plotting Univariate Distribution with distplot()

sns.distplot(tips['total_bill'])

#dist
#hist
#boxplot
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'],hist=False)
```

```python
#Plotting Bivariate Distribution with jointplot()

sns.jointplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, height=4, aspect=3)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips, height=4, aspect=3)
```

```python
tips["size"].value_counts()
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips, height=6, aspect=2)
```

```python
sns.relplot(x="total_bill", y="tip", style="size", hue="sex", data=tips, height=5, aspect=2)
```

```python
sns.relplot(x="total_bill", y="tip", style="time", hue="smoker", data=tips, size="size", height=5, aspect=2)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", col="sex", data=tips, height=5, aspect=1)
```

```python
sns.countplot(x="sex", data=tips, palette="husl")   #palette-color
```

```python
sns.catplot(x="day", y="total_bill", data=tips, height=5, aspect=2.5)
```

```python
sns.catplot(x="day", y="tip", data=tips, height=5, aspect=2.5, hue="sex")
```
